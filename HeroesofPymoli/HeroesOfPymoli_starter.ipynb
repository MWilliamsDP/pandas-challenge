## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins and group labels
bins = [0,10,15,20,25,30,35,40,1000]
groupNames =['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
#Create age group
data['Age Group']=pd.cut(data['Age'], bins, labels = groupNames, include_lowest = True)
#Groupby
AgeCategory = data.groupby('Age Group')
#Count
TotalCount = AgeCategory['SN'].nunique()
#Find percent
PercentbyAge=(TotalCount/players)*100
#Put into DF
AgeDemo = pd.DataFrame({'Percentage of Players': PercentbyAge, 'Total Count': TotalCount})
#Attempt format
AgeDemo.style.format({"Percentage of Players":"{:,.2f}%"})

AgeDemo
```

```python
# Group by player ID
playerSpend = data.groupby('SN')
#Count Purchases
Purchases = playerSpend['Purchase ID'].count()
#Calculate average
AVG=playerSpend['Price'].mean()
#Calculate total
ToT=playerSpend['Price'].sum()
#Create DF
Highrollers =  pd.DataFrame({'Purchase Count': Purchases, 'Average Purchase Price': AVG, 'Total Purchase Value': ToT}).sort_values(['Total Purchase Value'], ascending=False)
#CreateDF with highest spenders
#HighRollers = Spenders.sort_values(['Total Purchase Value'], ascending=False)
HighRollers.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
PopItems1=data[['Item ID', 'Item Name', 'Price']]
PopItems2=PopItems1.groupby(['Item ID', 'Item Name'])
TotItemsPurchase = PopItems2['Price'].count()
AvgPriceItems = PopItems2['Price'].mean()
TotPurchaseValue = PopItems2['Price'].sum()

PopItemsDF=pd.DataFrame({'Purchase Count': TotItemsPurchase, 'Item Price': AvgPriceItems, 'Total Purchase Value':TotPurchaseValue}).sort_values(['Purchase Count'], ascending=False)
PopItemsDF.head(10)
```
